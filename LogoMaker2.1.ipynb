Nikonバスターの機能を付ける


```python
import os
import sys
import shutil
from enum import Enum, auto
from tkinter import filedialog
from tkinter import messagebox as tkMessageBox #python3
from tkinter import filedialog as tkFileDialog #python3
import tkinter, tkinter.filedialog
from PIL import Image

root = tkinter.Tk()
root.withdraw()

#元画像フォルダー選択
msg = '画像フォルダーを選択してください。'
img_dir_path = tkinter.filedialog.askdirectory(title=msg)
if (not img_dir_path): #[キャンセル]クリック時の処理
    print('フォルダーを選んでください。')
    sys.exit()

output_dir_path = img_dir_path

all_pictures = os.listdir(img_dir_path)


message = 'ロゴ画像を選択してください。'
logo_path = tkFileDialog.askopenfilename(title=message) 

corner_X = input('ロゴ位置X(中央に挿入する場合は0と入力): ')#この値をイベントごとに毎回変える
corner_Y = input('ロゴ位置Y: ')#この値をイベントごとに毎回変える


WM = Image.open(logo_path)
w_wm, h_wm = WM.size
WM_1 = WM
WM_3 = WM.rotate(180 , expand=True)
WM_6 = WM.rotate(90 , expand=True)
WM_8 = WM.rotate(270 , expand=True)

def find_all_files(directory):#ディレクトリ以下全てをリスト化する定義
    for root, dirs, files in os.walk(directory):
        yield root
        for file in files:
            yield os.path.join(root, file)


class Mode(Enum):
    BLG = auto() #
    BLG2 = auto() #

def make_img(img, img_name, mode, output_path,orientation):
    #MAX_W_BLG = 3600 #画像の幅の上限※こちらはあまり意味のない数値です
    #MAX_H_BLG = 3600 #画像の高さの上限　
       
    #  サイズを設定
    #w, h = MAX_W_BLG, MAX_H_BLG
    w_img, h_img = img.size
    
    img = img.resize((3600,2400))
        
    #リサイズ
    #img.thumbnail((w, h))
    #画像を付加
    w_img, h_img = img.size
  
    if int(corner_X) > 0 and orientation == 1 :
        img.paste(WM_1, (w_img - w_wm - int(corner_X), h_img - h_wm-int(corner_Y)),WM_1)#横
    elif int(corner_X) > 0 and orientation == 6 :
        img.paste(WM_6, (w_img - h_wm- int(corner_Y), int(corner_X)),WM_6)#縦時計回り
    elif int(corner_X) > 0 and orientation == 8 :
        img.paste(WM_8, (int(corner_Y),h_img-w_wm - int(corner_X)),WM_8)#縦時計回り
        
    if int(corner_X) == 0 and orientation == 1 :
        img.paste(WM_1, (int(w_img*0.5) - int(w_wm*0.5) , h_img - h_wm-int(corner_Y)),WM_1)#横
    elif int(corner_X) == 0 and orientation == 6 :
        img.paste(WM_6, (w_img - h_wm- int(corner_Y), int(h_img*0.5) - int(w_wm*0.5)),WM_6)#
    elif int(corner_X) == 0 and orientation == 8 :
        img.paste(WM_8, (int(corner_Y),int(h_img*0.5) - int(w_wm*0.5)),WM_8)#縦時計回り

    #ファイル名に文字列を付加して保存
    fname, ext = os.path.splitext(img_name)
    img.save(os.path.join(output_path, fname + ext), quality=95, dpi=(300, 300), exif=img.info['exif'])
def make_img_noexif(img, img_name, mode, output_path):
    #MAX_W_BLG = 3600 #画像の幅の上限
    #MAX_H_BLG = 3600 #画像の高さの上限　
    w_img, h_img = img.size
    
    if w_img > h_img:
        img = img.resize((3600,2400))
        
    elif w_img < h_img:
        img = img.resize((2400,3600))
        
    #リサイズ
    #img.thumbnail((w, h))
    #画像を付加
    w_img, h_img = img.size
  
    if int(corner_X) > 0 :
        img.paste(WM_1, (w_img - w_wm - int(corner_X), h_img - h_wm-int(corner_Y)),WM_1)
    
    elif int(corner_X) == 0 :
        img.paste(WM_1, (int(w_img*0.5) - int(w_wm*0.5) , h_img - h_wm-int(corner_Y)),WM_1)
    
    
    #ファイル名に文字列を付加して保存
    fname, ext = os.path.splitext(img_name)
    img.save(os.path.join(output_path, fname + ext), quality=95, dpi=(300, 300), exif=img.info['exif'])
def make_img_lostexif(img, img_name, mode, output_path):
    #MAX_W_BLG = 3600 #画像の幅の上限
    #MAX_H_BLG = 3600 #画像の高さの上限　
    w_img, h_img = img.size
    
    if w_img > h_img:
        img = img.resize((3600,2400))
        
    elif w_img < h_img:
        img = img.resize((2400,3600))
        
    #リサイズ
    #img.thumbnail((w, h))
    #画像を付加
    w_img, h_img = img.size
  
    if int(corner_X) > 0 :
        img.paste(WM_1, (w_img - w_wm - int(corner_X), h_img - h_wm-int(corner_Y)),WM_1)
    
    elif int(corner_X) == 0 :
        img.paste(WM_1, (int(w_img*0.5) - int(w_wm*0.5) , h_img - h_wm-int(corner_Y)),WM_1)
    
    
    #ファイル名に文字列を付加して保存
    fname, ext = os.path.splitext(img_name)
    img.save(os.path.join(output_path, fname + ext), quality=95, dpi=(300, 300))

# 元画像フォルダー内のファイル1つずつ処理

#jpgとJPGのみをリスト化
all_picture_path = []
pic = []

for file in find_all_files(img_dir_path):#全ファイルとフォルダをリスト化
    all_picture_path.append(file)
            
            
for file in all_picture_path:#全ファイルとフォルダのリストの中から.jpgを別リストに移動
    base, ext = os.path.splitext(file)
    if ext == '.jpg':
        pic.append(file)

for file in all_picture_path:#全ファイルとフォルダのリストの中から.JPGを別リストに移動
    base, ext = os.path.splitext(file)
    if ext == '.JPG':
        pic.append(file)      

#print(pic)

i= 0
while i < len(pic):
    img_path = os.path.join(img_dir_path, pic[i])
    img_pil = Image.open(img_path)

    #ファイルの移動先フォルダー作成
    output_path = pic[i]
    
    #画像を作成
    exif = img_pil._getexif()
    str_exif = str(exif)#検索のために文字列に変換
    
    #exifからAdobeソフトを使ったか否かを判定
    if 'Adobe' in str_exif and len(str_exif) > 4:#Adobeソフトで編集後の画像の処理
        make_img_noexif(img_pil.copy(), pic[i], Mode.BLG,output_path)

    elif 'Adobe' not in str_exif and len(str_exif) == 4:#Exifが完全にない画像の処理
        make_img_lostexif(img_pil.copy(), pic[i], Mode.BLG,output_path)
        
    elif 'Adobe' not in str_exif and len(str_exif) > 4:#編集していない画像の処理
        orientation = exif.get(0x112, 1)
        make_img(img_pil.copy(), pic[i], Mode.BLG,output_path,orientation)
        
 
    
    #元画像(PIL)を閉じる
    img_pil.close()
    percent = (int(i+1)/int(len(pic)))
    print("\r"+'進捗'+str(round(100 * percent,1)) ,end=''+'％')
    i = i + 1
    
pic = []#一旦空にする
print('完了')
```
